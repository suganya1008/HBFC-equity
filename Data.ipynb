```python
import pandas as pd
```

```python
df["Date "]=pd.to_datetime(df["Date "],format="%d-%b-%Y")
```

```python
df.drop(columns=["series "],axis=1,inplace=True)
```

```python
for i in df.select_dtypes(include=["object"]).columns:
  df[i]=df[i].apply(lambda i:float(i.replace(",", "")))
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from scipy import stats
```

```python
for i in range(0,35):
  sample1 = df["vwap "].sample(frac = 0.04)
  sample2 = df["52W H "].sample(frac = 0.03)
  t_test, P_value = stats.ttest_ind(sample1,sample2)
```

```python
P_value
```

```python
t_test
```

```python
for i in range(0,31):
  sample1  = df['vwap '].sample(frac = 0.04)
  sample2  = df['52W H '].sample(frac = 0.03)
  t_test, p_value = stats.ttest_ind(sample1,sample2)
  print(t_test,p_value)
  if p_value < 0.05:
    print("h0 rejected,ha accepted, data is independent in nature")
  else:
    print("ha rejected,ho accepted, data is dependent in nature")
```
